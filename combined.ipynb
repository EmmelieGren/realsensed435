# Frågeställning: Kan man kombinera färgbilden med djupbilden för att få bätter resultat vid träning av Yolo modell.
- Notebook för att göra kombinationer av bilderna.
- Totalt 23 bilder med påde färgbild samt djupbild.

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import os
```

# Test 1 
- Kombinerar bilderna procentuellt

```python
color_img_path = "1_color.png"
depth_img_path = "1_depth.png"

color_img = cv2.imread(color_img_path)
depth_img = cv2.imread(depth_img_path, cv2.IMREAD_GRAYSCALE)
```

```python
print(depth_img.shape)
print(color_img.shape)
```

```python
combined_img = 0.7 * color_img + 0.3*depth_img[:,:,np.newaxis]
```

```python
combined_img.shape
```

### Hur kombineringen ser ut:

```python
plt.imshow(combined_img/ 255)
```

### Göra alla i foldern

```python
output_folder = 'combined'

for i in range(1, 25):
    color_img_path = f"img/{i}_color.png"
    depth_img_path = f"img/{i}_depth.png"


    if os.path.exists(color_img_path) and os.path.exists(depth_img_path):
        color_img = cv2.imread(color_img_path)
        depth_img = cv2.imread(depth_img_path, cv2.IMREAD_GRAYSCALE)

        # tryck ihop dem
        combined_img = 0.7 * color_img + 0.3 * depth_img[:, :, np.newaxis]

        combined_img_path = os.path.join(output_folder, f"{i}_combined.png")
        cv2.imwrite(combined_img_path, combined_img)

        print(f"Bild sparad: {combined_img_path}")
    else:
        print(f"Hittade inte bild nr:{i}")
```

# Test 2
- Ersätt ett färglager med djupbild.

```python
color_img_path = "1_color.png"
depth_img_path = "1_depth.png"

color_img = cv2.imread(color_img_path)
depth_img = cv2.imread(depth_img_path, cv2.IMREAD_GRAYSCALE)
```

```python
depth_img.shape
```

```python
new_img = np.copy(color_img)

new_img[:, :, 2] = depth_img # ersätteren färg med depth

print(new_img.shape)
```

### Exempel på hur bilden ser ut:

```python
plt.imshow(new_img/ 255)
```

### Gör alla i foldern

```python
output_folder = 'combined2'

for i in range(1, 25):
    color_img_path = f"img/{i}_color.png"
    depth_img_path = f"img/{i}_depth.png"


    if os.path.exists(color_img_path) and os.path.exists(depth_img_path):
        color_img = cv2.imread(color_img_path)
        depth_img = cv2.imread(depth_img_path, cv2.IMREAD_GRAYSCALE)

        print(color_img.shape)
        print(depth_img.shape)

        combined_img = np.copy(color_img)
        # ersätt en färg med djup
        combined_img[:, :, 1] = depth_img

        combined_img_path = os.path.join(output_folder, f"{i}_combined.png")
        cv2.imwrite(combined_img_path, combined_img)

        print(f"Bild sparad: {combined_img_path}")
    else:
        print(f"Hittade inte bild nr:{i}")
```
